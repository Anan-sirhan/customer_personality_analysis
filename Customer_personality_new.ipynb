```python

marital_status_counts = df['Marital_Status'].value_counts().reset_index()
marital_status_counts.columns = ['Marital_Status', 'Number of customers']

plt.figure(figsize=(10,5))
ax = sns.barplot(data=marital_status_counts, x='Marital_Status', y='Number of customers', palette='pastel')

for index, row in marital_status_counts.iterrows():
    ax.text(index, row['Number of customers'] + 1,
            f"{int(row['Number of customers'])}",
            color='black', ha="center", fontsize=12)

plt.show()
```

```python
# Check for duplicate columns
if df.columns.duplicated().any():
    print("Warning: DataFrame contains duplicate column names. Dropping duplicates.")
    # Keep the first occurrence of each column name
    df = df.loc[:, ~df.columns.duplicated(keep='first')]

# Convert the Customer_joining_date column to datetime format
df['Customer_joining_date'] = pd.to_datetime(df['Customer_joining_date'], errors='coerce')

# Create a new column with the quarter
df['Join_Quarter'] = df['Customer_joining_date'].dt.to_period('Q').astype(str)

# Ensure there are no missing values in the quarter column that might have resulted from conversion errors
df = df.dropna(subset=['Join_Quarter'])

# Count customers by quarter
quarter_counts = df.groupby('Join_Quarter').size().reset_index(name='Number_of_customers')

# Sort by quarter (convert to Period again for chronological order)
quarter_counts['Join_Quarter'] = pd.PeriodIndex(quarter_counts['Join_Quarter'], freq='Q')
quarter_counts = quarter_counts.sort_values('Join_Quarter')

# Convert back to quarter strings for plotting
quarter_counts['Join_Quarter'] = quarter_counts['Join_Quarter'].astype(str)

# Plot the graph
plt.figure(figsize=(12,6))
sns.lineplot(data=quarter_counts, x='Join_Quarter', y='Number_of_customers', marker='o')

plt.title('Number of Customers Joined per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df_merged = pd.merge(df_customers, df_spending, on='Customer_id')

# סינון לפי הכנסה
filtered_data = df_merged[df_merged['Income'] < 250000]

# ציור הגרף
sns.jointplot(x='Income', y='Amount_premium_products', data=filtered_data, kind='reg')

plt.show()
```

```python


# List of columns to exclude

exclude_cols = ['Z_CostContact', 'Z_Revenue', 'Year_Birth']

# Filter only numerical columns, excluding the specified ones

numeric_df = df.select_dtypes(include='number').drop(columns=exclude_cols, errors='ignore')

# Compute the correlation matrix
corr = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(18,12))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, square=True)

plt.title('Heatmap of Numerical Feature Correlations (Excluding Z_CostContact, Z_Revenue, Year_Birth)')
plt.tight_layout()
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
```

```python
#Average Total Spending by Marital Status and Income Group
#This analysis explores how customers' total spending varies by their marital status and income group.
#We use a heatmap to visualize patterns in spending behavior.

```

```python

df_spending["Total_Spending"] = df_spending[
    ['Amount_wines', 'Amount_fruits', 'Amount_meat_products',
     'Amount_fish_products', 'Amount_sweet_products', 'Amount_premium_products']
].sum(axis=1)

spending_summary = df_spending[["Customer_id", "Total_Spending"]]
customer_summary = df_customers[["Customer_id", "Marital_Status", "Income"]]

customer_summary["Income_Group"] = pd.qcut(customer_summary["Income"], 4, labels=["Low", "Medium", "High", "Very High"])


temp = pd.merge(customer_summary, spending_summary, on="Customer_id")


heatmap_data = temp.groupby(["Marital_Status", "Income_Group"])["Total_Spending"].mean().unstack()


plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="BuGn")
plt.title("Average Total Spending by Marital Status and Income Group")
plt.xlabel("Income Group")
plt.ylabel("Marital Status")
plt.tight_layout()
plt.show()
```
